### Dataset for Funnel Analysis

```python
df_Events = excel_data.parse(excel_data.sheet_names[0])
```

```python
df_Events.head()
```

```python
df_Events.info()
```

```python
df_Events['created_date'] = pd.to_datetime(df_Events['created_at']).dt.date
df_Events.groupby(['created_date', 'event_type', 'gender']).user_id.nunique()
```

```python
# Aggregate
df_Funnel = df_Events.groupby(['created_date', 'event_type', 'gender'])['user_id'].nunique().reset_index()

# Rename
df_Funnel.rename(columns={'user_id': 'unique_user_count'}, inplace=True)

df_Funnel.head()
```

```python
df_Funnel.to_excel('Funnel_table.xlsx')
```

### Dataset for Cohort Analysis

```python
df_Sales = excel_data.parse(excel_data.sheet_names[1])
```

```python
df_Sales.head()
```

```python
# Aggregate created_at column from Events to Sales

agg_events = df_Events.groupby('user_id', as_index=False).agg(min_date=('created_at', 'min'))
output_table = pd.merge(df_Sales, agg_events, on='user_id', how='left')
```

```python
output_table.head()
```

```python
output_table.info()
```

```python
output_table['min_date'] = pd.to_datetime(output_table['min_date'], errors='coerce')
output_table.info()
```

```python
output_table['min_date'] = output_table['min_date'].dt.date
output_table.info()
```

```python
output_table['min_date'] = pd.to_datetime(output_table['min_date'], errors='coerce')
output_table.info()
```

```python
output_table['cohort_date'] = output_table['order_date'] - output_table['min_date']
output_table['cohort_date'] = output_table['cohort_date'].dt.days
```

```python
output_table.head()
```

```python
output_table.info()
```

```python
output_table.to_excel('Cohort_table.xlsx')
```
