# Datetime Exploration

Bagaimana total pendapatan kotor pada minggu pertama?

````python
df["Date"] = pd.to_datetime(df["Date"],format="%m/%d/%Y")
````

````python
df.info()
````

````python
df['Time'] = df['Time'].str.strip()
````

````python
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M', errors='coerce').dt.time
````

````python
# Ekstraksi Hari dalam Seminggu
df['DayOfWeek'] = df['Date'].dt.day_name()

# Ekstrak Bulan
df["Month"] = df["Date"].dt.month

# ekstrak tahun
df["year"] = df["Date"].dt.year
````

````python
df.info()
````

````python
df.head()
````

````python
def get_week_of_month(dt):
  day_of_month = dt.day
  if day_of_month >= 29:
    week_of_month = 1
  else:
    week_of_month = (day_of_month - 1) // 7 + 1
  return week_of_month
````

````python
df["Week"] = df["Date"].apply(get_week_of_month)
````

````python
df.head()
````

````python
df.groupby("Week")["gross income"].describe()
````

```Insight```

- Untuk data ini pada awalnya ada week ke 5 yang frekuensinya tidak terlalu banyak, sehingga saya memutukan untuk menggabungkannya ke dalam week 1 yang di sini akan saya definisikan sebagai awal bulan.
- Dari sini terlihat bahwa ternyata peak pendapatan kotor tertinggi itu berada di minggu ketiga setiap bulannya.
- Berdasarkan data ini maka bisa diadakan semacam promo tertentu atau pun mempertimbangkan untuk merilis produk-produk baru di minggu ketiga.

# 2. Standar Scaller for Exploratory Data Analysis ?
`clue` : lakukan tanpa splitting dataset ke train dan test
- lakukan pada kolom Total, Unit Price, dan tambahkan 2 kolom lainnya
- dan tulis insight dari yang anda dapat dari grafik tersebut

````python
from sklearn.preprocessing import StandardScaler
columns_to_stdscale = ["Total", "Unit price", "gross income", "Quantity"]

scaler = StandardScaler()
df[["Total_stds", "Unit price_stds", "gross income_stds", "Quantity_stds"]] = scaler.fit_transform(df[columns_to_stdscale])
````

````python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Sebelum Standar Scaler
ax1.set_title("Sebelum Standar Scaler")
sns.kdeplot(df["Total"], ax=ax1, label="Total", color='blue', fill=True, alpha=0.2)
sns.kdeplot(df["Unit price"], ax=ax1, label="Unit price",  color='orange', fill=True, alpha=0.2)
sns.kdeplot(df["gross income"], ax=ax1, label="gross income",  color='green', fill=True, alpha=0.2)
sns.kdeplot(df["Quantity"], ax=ax1, label="Quantity",  color='red', fill=True, alpha=0.2)

# Setelah Standar Scaler
ax2.set_title("Setelah Standar Scaler")
sns.kdeplot(df["Total_stds"], ax=ax2, label="Total (scaled)",  color='blue', fill=True, alpha=0.2)
sns.kdeplot(df["Unit price_stds"], ax=ax2, label="Unit price (scaled)",  color='orange', fill=True, alpha=0.2)
sns.kdeplot(df["gross income_stds"], ax=ax2, label="gross_income (scaled)",  color='green', fill=True, alpha=0.2)
sns.kdeplot(df["Quantity_stds"], ax=ax2, label="Quantity (scaled)",  color='red', fill=True, alpha=0.2)

ax1.set_xlabel('Rentang Nilai Sebelum Feature Scalling (Standar Scaller)')
ax2.set_xlabel('Rentang Nilai Setelah Feature Scalling (Standar Scaller)')

ax1.legend()
ax2.legend()

plt.tight_layout()
plt.show()
````

````python
df[["Total","Total_stds"]].head()
````

````python
df[["Total","Total_stds"]].describe()
````

```Insight```

Dari kedua charts tersebut sangat jelas terlihat bedanya di mana sebelum dilakukan Standard Scaling distribusi persebaran datanya terlihat sangat ekstrim sehingga sulit untuk dibandingkan. Sedangkan sesudah Standard scaling semua nilai memiliki skala dari 0 sampai 1 sehingga lebih mudah untuk dibuat permodelan Machine Learning dan algoritmanya.

Dalam grafik setelah Standard scaling terlihat bahwa total dan gross Income memiliki persebaran positive skewed, sedangkan unit price dan quantity memiliki persebaran multimodal.

# 3. MinMax Scaller for Exploratory Data Analysis ?
`clue` : lakukan tanpa splitting dataset ke train dan test
- lakukan pada kolom Total, Unit Price, dan tambahkan 2 kolom lainnya
- dan tulis insight dari yang anda dapat dari grafik tersebut

````python
from sklearn.preprocessing import MinMaxScaler

columns_to_minmaxscale = ["Total", "Unit price", "gross income", "Quantity"]

scaler = MinMaxScaler()
df[["Total_minmax", "Unit price_minmax", "gross income_minmax", "Quantity_minmax"]] = scaler.fit_transform(df[columns_to_minmaxscale])
````

````python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# Sebelum MinMax scaling
ax1.set_title("Sebelum MinMax scaling")
sns.kdeplot(df["Total"], ax=ax1, label="Total", color='blue', fill=True, alpha=0.2)
sns.kdeplot(df["Unit price"], ax=ax1, label="Unit price",  color='orange', fill=True, alpha=0.2)
sns.kdeplot(df["gross income"], ax=ax1, label="gross income",  color='green', fill=True, alpha=0.2)
sns.kdeplot(df["Quantity"], ax=ax1, label="Quantity",  color='red', fill=True, alpha=0.2)

# Sesudah MinMax scaling
ax2.set_title("Setelah MinMax scaling")
sns.kdeplot(df["Total_minmax"], ax=ax2, label="Total (scaled)",  color='blue', fill=True, alpha=0.2)
sns.kdeplot(df["Unit price_minmax"], ax=ax2, label="Unit price (scaled)",  color='orange', fill=True, alpha=0.2)
sns.kdeplot(df["gross income_minmax"], ax=ax2, label="gross_income (scaled)",  color='green', fill=True, alpha=0.2)
sns.kdeplot(df["Quantity_minmax"], ax=ax2, label="Quantity (scaled)",  color='red', fill=True, alpha=0.2)

ax1.set_xlabel('Rentang Nilai Sebelum Feature Scalling (MinMax Scaller)')
ax2.set_xlabel('Rentang Nilai Setelah Feature Scalling (MinMax Scaller)')

ax1.legend()
ax2.legend()

plt.tight_layout()
plt.show()
````

````python
df[["Total","Total_minmax"]].head()
````

````python
df[["Total","Total_minmax"]].describe()
````

```Insight```

Kurang lebih sama dengan Standard scaling, namun bedanya di Min-Max scaling skala nilainya tidak hanya 0 sampai 1.
